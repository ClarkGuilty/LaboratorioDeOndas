```python
#Importar las librerías numpy y matplotlib.
%pylab inline
import math
```

```python
#Cargar la información de péndulo de bronce
data = loadtxt('pendulo1.csv',delimiter=';')
x = data[:,0]
y = data[:,1]
scatter(x,y)
```

```python
#Se escribe la ecuación que, de acuerdo al modelo clásico del péndulo, relaciona la longitud del péndulo con su periodo.
def per(x, g):
    return 12.0*math.pi*sqrt(x/g)
```

```python
#Se hace finalmente la regresión 
from scipy.optimize import curve_fit


g = curve_fit(per,x,y, p0= 1000)[0][0]
```

```python
y_fit = per(x, g)
g
```

```python
# -*- coding: utf-8 -*-
plot(x,y_fit/6.0, color = 'red', label = 'Regresion, g = 9.63 m/s^2')
scatter(x,y/6.0, label = 'Datos')
xlabel("Largo del pendulo.")
ylabel("Periodo de una oscilacion del pendulo")

legend(loc=0)
```

```python
dos = loadtxt('tres.txt',dtype = None, delimiter=' ')
```

```python
a = dos[:,0]
b = dos[:,1]
```

```python
def ex(x, q,w):
    return q*e**(w*x)
heh = curve_fit(ex,a,b)
q = heh[0][0]
w = heh[0][1]
```

```python
b_fit = ex(a,q,w)
```

```python
plot(a,b_fit, color = 'red', label = 'Regresion')
scatter(a,b, label = 'Datos')
plot(a,b)
xlabel("Numero de oscilaciones. (nT)")
ylabel("Amplitud angular. (Grados)")

legend(loc=0)
```
